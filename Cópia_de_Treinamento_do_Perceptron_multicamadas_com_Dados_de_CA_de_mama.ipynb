```python
from sklearn.preprocessing import LabelEncoder  # A "sklearn.proprocessing" vai transformar as variáveis categoricas em números.

cols = ['diagnosis'] # É nessesário descrever quais as colunas que tem váriaveis categoricas para serem modificadas.
le = LabelEncoder()
dataset[cols] = dataset[cols].apply(le.fit_transform)

```

```python
x = dataset.iloc[:, 2:] # 1: pegar da coluna 1 em diante, :-1 pegar todas as colunas menos a última.
y = dataset.iloc[:, 1]

x
```

```python
y  #testando o "x" e "y"
```

```python
# Vai utilizar a função "sklearn" para separar os dados de treino e teste
from sklearn.model_selection import train_test_split #importando os dados para treino
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2) #separando o "x" e "y" e definindo que apenar 0.2, ou seja, 20% dos dados serão usados para teste, o restante para treino.
```

```python
#NORMALIZAÇÃO, cada coluna tem sua propria estrutura númerica, sendo necessário normalizar para que todas tenha o mesmo "valor númerico".
from sklearn.preprocessing import StandardScaler #vamos usar o StandarScaler para normalizar tudo.
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

x_test
```

```python
y_test
```

#Arquitetura de Rede Neural (MLP) 
Trabalhar diretamente com o "tf", ou seja, tensorFlow tem códigos muito complicados de mexer, por isso vamos usar o "Keras" junto para facilitar

O "models-sequential" é justamente para chamar um rede neural com propragação pra frente (tradicinal).
Na rede neural abaixo: temos 2 camadas ocultas, com "dense", ou seja, todos os neurônios estão conectados com as camadas da frente.
Cada camada tem "units=6", ou seja, 6 neurônios.
A função de ativação que vamos usar é a "relu", que é praticamente uma função linear e funciona muito bem no treinamento da rede neural.
A ultima camada, com apenar 1 neurônio irá me dá a resposta, se o CA é banigno ou maligno.Esse neuronio tem uma função binária, para sim ou não, e por isso vamos usar a função sigmoid.
O otimizador que vamos usar é o "adam"
O relatório de perda, para saber que está acertando ou não será dado pelo "binary_crossentropy".
Na acucácia vamos usar "accuracy".

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# TESTANDO A REDE
- O "batch-size" seleciona um quantidade de dados que seram selecionados para ajustar a rede, vou usar o número 32. Mas, poderia ser 16,32,64...
- O "epochs" é quantidade de vezes que será o ajuste com os 32 dados aleatórios para ajustar a rede.

```python
ann.fit(x_train, y_train, batch_size=32, epochs=50)
```

#Resultados
Quando usamos a função "ann.predict(x_test)", estamos jogando o conjunto de teste para a rede neural, pois ele é o objeto da rede neural já treinado.

O "y_pred" são os resultados, porém eles vão aparecer em valor número, se o valor for proximo de 0, provavelmente a chance de ter um CA maligno é pouca, agora se for proxima de 1 maior é a chance.
Por causa disso, vamos usar o código "y_pred > 0.5" para ele analisar os número é categorizar em <0,5 risco de CA maligno, >0,5 baixo risco de CA maligno.

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)

pred_array = 1 * y_pred.reshape(len(y_pred), 1) # O que a rede préve, é feita a multiplicação por 1 para formação de número binário (pois 0x1=0 e 1x1=1 e assim dá pra aferir se tem risco ou não de AVC)
test_array = y_test.values.reshape(len(y_test), 1) # O que realmente é verdade

np.concatenate([pred_array, test_array], axis=1) # Aqui vamos concatenar os dados para descobrir a efetividade do treinamento
```

Visualização dos resultados utilizando na matriz de confusão

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred) 
#vn, fp(rede disse que é 1, mas é 0), fn (rede disse que é 0, mas é 1), vp

cm
```
